```python
import pandas as pd
import sqlalchemy as sa
#!pip install sqlalchemy psycop2
```

```python
engine = sa.create_engine('postgresql://root:root@localhost:5432/ny_taxi')
engine.connect()
```

```python
df = pd.read_parquet('yellow_tripdata_2021-01.parquet')
#get the DDL schema statement definition in postgress sql dialect
print(pd.io.sql.get_schema(df, name='yellow_taxi', con=engine))
```

```python
df.to_sql('yellow_taxi', con=engine, chunksize=100000, if_exists='append')
```

```python
#check data is there
pd.read_sql('select  *  from yellow_taxi', con=engine).head(100)
```
